```python
# Exercise 1

def GenPet(n, k, check=False):

    G = Graph(2*n)
    for i in range(n):
        G.add_edge(i, (i+1)%n)
        G.add_edge(i, i+n)
        if i+k >= n:
            G.add_edge(i+n, (i+k)%n + n)
        else:
            G.add_edge(i+n, i+n+k)
    if check == True:
        H = graphs.GeneralizedPetersenGraph(n, k)
        if G.is_isomorphic(H):
            print("The graph is isomorphic to Generalized Petersen Graph")
        else:
            print("The graph is not isomorphic to Generalized Petersen Graph")
        
        
    return G
GenPet(7, 2, True).show()
GenPet(4, 1).show()
```

```python
#Exercise 2

def Kneser(n, k, check=False):
    
    if not 0 <= k <= n:
        return "Your input is not valid."
    
    G = Graph()
    for i in Subsets(n, k):
        G.add_vertex(i)
    
    for i in G.vertices():
        for j in G.vertices():
            if i.intersection(j) == Set([]) and not G.has_edge(i, j):
                G.add_edge(i, j)
                
    if check:
        H = graphs.KneserGraph(n, k)
        return G.is_isomorphic(H)
    
    return G


Kneser(5, 2)
```
